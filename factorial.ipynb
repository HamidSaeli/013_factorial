# Project 13: Calculate Factorial of a Number

```python
# This project teaches loops and multiplication
```

```python
# Step 1: Ask the user to enter a number
num = int(input("Enter a positive integer: "))
```

```python
# Step 2: Set the initial factorial value
factorial = 1
```

```python
# Step 3: Multiply all numbers from 1 to num
for i in range(1, num + 1):
    factorial *= i   # same as factorial = factorial * i
```

```python
# Step 4: Print the result
print(f"The factorial of {num} is: {factorial}")
```
